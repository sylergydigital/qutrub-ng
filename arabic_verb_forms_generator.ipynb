# Arabic Verb Forms Generator (I-X)

This notebook generates all 10 Arabic verb forms from a given root verb.

## The 10 Arabic Verb Forms:

1. **Form I (Ø§Ù„ÙØ¹Ù„ Ø§Ù„Ù…Ø¬Ø±Ø¯)**: The basic or original root form
2. **Form II (ÙÙØ¹ÙÙ‘Ù„Ù)**: Often conveys causative or intensive meaning
3. **Form III (ÙÙØ§Ø¹ÙÙ„Ù)**: Usually indicates interaction or reciprocity
4. **Form IV (Ø£ÙÙ’Ø¹ÙÙ„Ù)**: Typically has causative or declarative meanings
5. **Form V (ØªÙÙÙØ¹ÙÙ‘Ù„Ù)**: Generally a reflexive or intensive of Form II
6. **Form VI (ØªÙÙØ§Ø¹ÙÙ„Ù)**: Reflexive or reciprocal of Form III
7. **Form VII (Ø§ÙÙ†Ù’ÙÙØ¹ÙÙ„Ù)**: Often passive or reflexive
8. **Form VIII (Ø§ÙÙÙ’ØªÙØ¹ÙÙ„Ù)**: Reflexive or related to taking part in the action
9. **Form IX (Ø§ÙÙÙ’Ø¹ÙÙ„ÙÙ‘)**: Rare; usually relates to colors or physical defects
10. **Form X (Ø§ÙØ³Ù’ØªÙÙÙ’Ø¹ÙÙ„Ù)**: Implies seeking or asking for the action of the root

## Try Multiple Verbs
